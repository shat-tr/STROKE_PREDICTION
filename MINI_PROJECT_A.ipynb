```python
import numpy as np
df['avg_glucose_level'] = pd.to_numeric(df['avg_glucose_level'], errors='coerce')
df['stroke'] = np.where(df['avg_glucose_level'] > 95.0, 'yes', 'no')
print(df)
```

```python
print(df[['avg_glucose_level','stroke']].head())
```

```python
interval=[0.0,50.0,100.0,150.0,200.0,250.0,300.0]
label=['0-50','51-100','101-150','151-200','201-250','251-300']
df['avg_glucose_level']=pd.cut(df['avg_glucose_level'],
                     bins=interval,
                     labels=label,
                     include_lowest='True')
print(df.head())
```

```python
int_mapping={1:'Yes', 0:'No'}
df['heart_disease']=df['heart_disease'].replace(int_mapping)
df['hypertension']=df['hypertension'].replace(int_mapping)
df.head()
```
